The rise of health concerns has highlighted the importance of ensuring safety measures, and this project seeks to leverage machine learning to assist me mask detection.

- **With Mask**: Images of individuals wearing face masks.

The dataset contains:

The images are resized to 128x128 pixels and are labeled accordingly.

### 7. Train Test Split

```python
# Splitting the Data in Train and Test
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

### 8. Scaling the Data

```python
# scaling the data

X_train_scaled = X_train/255

X_test_scaled = X_test/255
```

```python
X_train[0]
```

```python
X_train_scaled[0]
```

### 9. Building the Model

```python
num_of_classes = 2

model = keras.Sequential()

model.add(keras.layers.Conv2D(32, kernel_size=(3,3), activation='relu', input_shape=(128,128,3)))
model.add(keras.layers.MaxPooling2D(pool_size=(2,2)))


model.add(keras.layers.Conv2D(64, kernel_size=(3,3), activation='relu'))
model.add(keras.layers.MaxPooling2D(pool_size=(2,2)))

model.add(keras.layers.Flatten())

model.add(keras.layers.Dense(128, activation='relu'))
model.add(keras.layers.Dropout(0.5))

model.add(keras.layers.Dense(64, activation='relu'))
model.add(keras.layers.Dropout(0.5))


model.add(keras.layers.Dense(num_of_classes, activation='sigmoid'))
```

```python
# compile the neural network
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['acc'])
```

```python
# training the neural network
history = model.fit(X_train_scaled, Y_train, validation_split=0.1, epochs=10)
```

### 10. Model Evaluation

```python
loss, accuracy = model.evaluate(X_test_scaled, Y_test)
print('Test Accuracy =', accuracy)
```

```python
h = history

# plot the loss value
plt.plot(h.history['loss'], label='train loss')
plt.plot(h.history['val_loss'], label='validation loss')
plt.legend()
plt.show()

# plot the accuracy value
plt.plot(h.history['acc'], label='train accuracy')
plt.plot(h.history['val_acc'], label='validation accuracy')
plt.legend()
plt.show()
```

```python

# Get the image path from user input
input_image_path = input('Path of the image to be predicted: ').strip()

# Print the file path for debugging
print(f"File path provided: {input_image_path}")

# Check if the file exists
if not os.path.exists(input_image_path):
    print("Error: The file does not exist. Please check the file path.")
else:
    # Load the image
    input_image = cv2.imread(input_image_path)

    # Verify if the image was loaded
    if input_image is not None:
        # Convert BGR to RGB for displaying correctly with Matplotlib
        input_image_rgb = cv2.cvtColor(input_image, cv2.COLOR_BGR2RGB)
        
        # Display the image using Matplotlib
        plt.imshow(input_image_rgb)
        plt.title('Input Image')
        plt.axis('off')  # Hide the axes
        plt.show()

        # Resize the image
        input_image_resized = cv2.resize(input_image, (128, 128))

        # Scale the image
        input_image_scaled = input_image_resized / 255.0

        # Reshape the image
        input_image_reshaped = np.reshape(input_image_scaled, [1, 128, 128, 3])

        # Predict the class of the image
        input_prediction = model.predict(input_image_reshaped)

        # Print the raw prediction probability
        print(f"Prediction probability: {input_prediction[0][0]}")

        # Determine if the person is wearing a mask based on the probability
        if input_prediction[0][0] > 0.5:
            print('The person in the image is wearing a mask')
        else:
            print('The person in the image is not wearing a mask')
    else:
        print("Error: Could not load image. Check the file path and file format.")
```
